```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
P = np.random.uniform(12, 1, 60*10)
```

Power time series

```python
plt.figure(figsize = (10,5))
plt.plot(P)
plt.xlabel("Time (min)")
plt.ylabel("Power (kW)")
```

Can find the energy consumption using the formula: 

$$E = \delta t \sum_{i=0}^n p_i$$ 

```python
E = []
for i in range(len(P)):
    if i < 60:
        E.append(1/30 * np.sum(P[i:i+30]))
    elif i + 60 > len(P):
        E.append(1/30 * np.sum(P[i-30:i]))
    else:
        E.append(1/60 * np.sum(P[i - 30: i + 30]))
E = np.array(E)
    
```

```python
plt.figure(figsize = (10,5))
plt.plot(P, 'r--', label = "Power (kW)" )
plt.plot(E, "k", label = "Energy (kWh)")
plt.xlabel("Time (min)")
plt.ylabel("Power and Energy")
plt.legend()
```
